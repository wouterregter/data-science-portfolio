## Scene Detection

Next, I define a find_scenes function so I can use it to extract the scenes of the movie trailers. Subsequently, I test find_scenes on some of the movie trailers to see if the scenes approximately match my own visual inspection. After testing different thresholds, I realize that the threshold level to get good results varies significantly among the movie trailers. Hence, I decided to shortly consider each movie trailer individually to pick a threshold level. I realize this is not the fastest way, but it does ensure that the scenes are reasonably accurate. After choosing the right threshold level I save the scenes in a folder.

```python
from scenedetect import VideoManager
from scenedetect import SceneManager
from scenedetect.detectors import ContentDetector
from scenedetect.scene_manager import generate_images

def find_scenes(video_path, threshold=30.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(
        ContentDetector(threshold=threshold))
    base_timecode = video_manager.get_base_timecode()
    video_manager.set_downscale_factor()
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager, show_progress=False)    
    return scene_manager.get_scene_list(base_timecode)
```

```python
import cv2

filename = 'videos3\\THE ROAD trailer.mp4'

scene_list = find_scenes(filename, threshold=30)

cap = cv2.VideoCapture(filename)

frames = []
shot_length = []

for start_time, end_time in scene_list:
    duration = end_time - start_time
    frame = (start_time.get_frames() + int(duration.get_frames() / 2))
    cap.set(cv2.CAP_PROP_POS_FRAMES,frame)
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frames.append(frame)
    shot_length.append(duration.get_seconds())
    
print(len(scene_list))
```

## Face and Emotion Detection

### Defining load_image_from_path

First I define a function to load an image from a path. 

```python
import io
import urllib
from PIL import Image, ImageOps

def load_image_from_path(image_path, target_size=None, color_mode='rgb'):
    pil_image = image.load_img(image_path, 
                               target_size=target_size,
                            color_mode=color_mode)
    return image.img_to_array(pil_image)
```

### Testing the Face and Emotion Detection Models

Next, I test the performance of the face and emotion detection models. In the codeblocks below I tested a few samples of scenes (by manually changing the code) to see if the faces and emotions are recognized correctly in the movie trailers. The models mostly extracted the right number of faces and the right emotions.

```python
import numpy as np
from tensorflow.keras.preprocessing import image

pre_image = load_image_from_path('scenes1_oxford/frame_37.jpg', color_mode='grayscale')
gray_image = np.squeeze(pre_image).astype('uint8')

face_classification = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') # load the classifier only once!

faces = face_classification.detectMultiScale(gray_image, 1.3, 5) # detect the faces 
n_faces = len(faces) # get the number of faces
print('Detected {} faces!'.format(n_faces))
```

```python
from tensorflow.keras.models import load_model

def apply_offsets(face_coordinates, offsets):
    """
    Derived from https://github.com/oarriaga/face_classification/blob/
    b861d21b0e76ca5514cdeb5b56a689b7318584f4/src/utils/inference.py#L21
    """
    x, y, width, height = face_coordinates
    x_off, y_off = offsets
    return (x - x_off, x + width + x_off, y - y_off, y + height + y_off)

emotion_classifier = load_model('fer2013_mini_XCEPTION.102-0.66.hdf5') # load this only once! (not in a loop)

EMOTION_OFFSETS = (0, 0)
INPUT_SHAPE_EMOTION = emotion_classifier.input_shape[1:3]

labels = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

for face_coordinates in faces:
    x1, x2, y1, y2 = apply_offsets(face_coordinates, EMOTION_OFFSETS) 
    face_img = gray_image[y1:y2, x1:x2] # only get the face
    face_img = cv2.resize(face_img, (INPUT_SHAPE_EMOTION))
    face_img = face_img.astype('float32') / 255.0 # pre-processing 
    face_img = face_img - 0.5 # pre-processing specific to the emotion classifier
    face_img = face_img * 2.0 # pre-processing specific to the emotion classifier
    face_img = np.expand_dims(face_img, 0) # batch of one
    face_img = np.expand_dims(face_img, -1) # pre-processing specific to the emotion classifier
    probas = emotion_classifier.predict(face_img)
    label = labels[np.argmax(probas[0])]
    emotionprobas = dict(zip(labels, probas[0])) # make a dictionary from the labels and the probabilities
    print('This person looks {}'.format(label))
    print(emotionprobas)
```

### Extracting Emotions from all Movie Trailers

Next, I extract the mean emotion proportions per period from the movie trailers. First, I loop over the periods, then over the trailers, then over the scenes in a list. For each trailer I accumulate all the emotions of all scenes. From this I calculate lists with percentage distributions of the emotions per movie trailer. Then I calculate the means of these proportions for each period per emotion.

```python
import os

face_classification = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
emotion_classifier = load_model('fer2013_mini_XCEPTION.102-0.66.hdf5')

paths = os.listdir()
period_paths = [i for i in paths if 'period' in i] # The periods each have their own folder, period1, period2, and period3
mean_period_proportions = {}

for period_path in period_paths: # First I loop over all the period folders
    
    trailer_proportions = {}
    trailer_index = 0
    
    for trailer_path in os.listdir(period_path): # Then I loop over all the movie trailer folders
        
        emotions = []
        
        for scene_path in os.listdir(period_path + '/' + trailer_path): # Then I loop over all the frames
            
            # Face Detection
            
            pre_image = load_image_from_path(period_path + '/' + trailer_path + '/' + scene_path, color_mode='grayscale')
            gray_image = np.squeeze(pre_image).astype('uint8')
            faces = face_classification.detectMultiScale(gray_image, 1.3, 5) # detect the faces 
            
            # Emotion Detection

            EMOTION_OFFSETS = (0, 0)
            INPUT_SHAPE_EMOTION = emotion_classifier.input_shape[1:3]
            labels = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
            for face_coordinates in faces:
                x1, x2, y1, y2 = apply_offsets(face_coordinates, EMOTION_OFFSETS) 
                face_img = gray_image[y1:y2, x1:x2] # only get the face
                face_img = cv2.resize(face_img, (INPUT_SHAPE_EMOTION))
                face_img = face_img.astype('float32') / 255.0 # pre-processing 
                face_img = face_img - 0.5 # pre-processing specific to the emotion classifier
                face_img = face_img * 2.0 # pre-processing specific to the emotion classifier
                face_img = np.expand_dims(face_img, 0) # batch of one
                face_img = np.expand_dims(face_img, -1) # pre-processing specific to the emotion classifier
                probas = emotion_classifier.predict(face_img)
                label = labels[np.argmax(probas[0])]
                emotionprobas = dict(zip(labels, probas[0])) # make a dictionary from the labels and the probabilities
                
            # Steps to get the mean emotion proportions per period
                
                # Save all the emotions to a list for each movie trailer (resets at a new movie trailer)
                emotions.append(label)
                
        # For each movie trailer in a period, calculate a list containing the percentage distribution 
        # per label and save the list to a dictionary (resets at a new period)
        trailer_proportions[trailer_index] = [emotions.count(lab) / len(emotions) for lab in labels]
        
        trailer_index = trailer_index + 1
        
    # Turn the dictionary into a dataframe which now contains the columns as movie trailers in one period
    # and the rows as the different labels
    trailer_proportions = pd.DataFrame(trailer_proportions)
        
    # Calculate the mean proportions of all trailers in a single period per label by subsetting the rows in the dataframe 
    # and taking the mean 
    mean_period_proportions[period_path] = [trailer_proportions.iloc[i].mean() for i in np.arange(len(labels))]
    
# Putting the mean period proportions into a dataframe and transposing it yields a df with mean emotion proportions
# per period (rows) per emotion (column)
mean_period_proportions_df = pd.DataFrame(mean_period_proportions).T
mean_period_proportions_df.columns = labels
mean_period_proportions_df.index = ["1920-1940","1960-1980","2000-2020"]
mean_period_proportions_df = mean_period_proportions_df.round(4)
mean_period_proportions_df
```

### Visualizing the Result

Subsequently, I plot the results in a stacked bar chart, which seemed most appropriate to me as the results of interest are proportions.

```python
plt.style.use('ggplot')
# Source Stacked Barchart: https://pstblog.com/2016/10/04/stacked-charts
mean_period_proportions_df.plot.bar(stacked=True, figsize=(7,5))
plt.ylabel('Mean Proportion')
plt.xlabel('Period')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

### Discussion

The results suggest that the proportion of sad, neutral and fear have slightly declined ( < 5%), disgust, surprise and angry have stayed approximately the same, and happy has increased noticeably (  10%) from period 1 to period 3 in the movie trailers. Furthermore, the results indicate that sad is the most dominant emotion in movie trailers over all periods.
